# Credit Default

Tarefa 1: Ler o arquivo com o pandas

```python
import pandas as pd

# realiza a leitura dos dados
dataset = pd.read_csv('../data/lv-credit-card-default-1000.csv',sep=',')
# carrega os dados em um dataframe
data = pd.DataFrame(dataset)
# exibe os dados para verificação
data.tail()
```

# Etapa 2: Transformando e selecionando os dados

Após a leitura dos dados é importante identificar quais os campos ordinais que devem ser traduzidos para valores numéricos para que sejam mais facilmente manipulados pelos algoritmos

```python
import numpy as np

# aplica um filtro para alterar dados não padronizados no caso o sexo
maskF = data['SEX'] == "F"
data.iloc[data[maskF].index, 2] = 2
maskF = data['SEX'] == "M"
data.iloc[data[maskF].index, 2] = 1
# deleta todos os registros que possuam algum valor NaN e altera o próprio dataframe
data.dropna(inplace = True)

data.tail()
```

Transformação dos dados: Realizando a normalização dos dados

```python
from sklearn import preprocessing

labelColumn = 'DEFAULTED'
# separa os dados dos rótulos
columns = data.columns.difference([labelColumn])
features = data[columns]
targets = data[labelColumn]
# normaliza os dados
nFeatures = preprocessing.normalize(features)
dataNorm = pd.DataFrame(nFeatures)
# remonta a columa de rótulo nos dados normalizados
# dataNorm.head()
```

# Etapa 3: Reduzindo a dimensionalidade do problema

Agora é o momento de estabelecer seus critérios de seleção dos campos. Você pode selecionar por métodos de filtragem (Valores faltantes, baixa variância, alta correlação), força bruta (BFE ou FFC) ou executar o PCA. O importante é encontrar um conunto de campos menor que o inicial para reduzir a complexidade computacional.

```python
#from sklearn.feature_selection import VarianceThreshold
from sklearn.decomposition import PCA

# exemplo de seleção por variância do campo
#selector = VarianceThreshold(0.0)
#dataF = selector.fit_transform(dataNorm)
# remonta o DataFrame
#dataFiltered = pd.DataFrame(dataF)
#dataFiltered.head()

# Create correlation matrix
corr_matrix = dataNorm.corr().abs()
# corr_matrix

# pca = PCA(n_components=3)
# pca.fit(dataNorm)
# dataNorm = pca.transform(dataNorm)
# dataNorm.values
```

Seleciona as colunas que devem ser eliminadas por um limiar

```python
# Seleciona somente o triangulo superior da matriz
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
#upper
# Encontra as colunas com variancia menor que 0.85
to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]
#to_drop
dataFiltered = dataNorm.drop(dataNorm.columns[to_drop], axis=1)
dataFiltered['DEFAULTED'] = targets;
dataFiltered.head()
```

# Etapa 4: Criar e treinar o modelo de classificação

Esse é o momento de escolher qual o algoritmo você quer usar. Usar mais de um (ou um método ensemble) ajuda a ter a noção de qual a melhor estratégia para o problema.

Esse também é o momento de criar as suas bases de treinamento, validação e teste. Realizar a validação cruzada também ajuda bastante e é muito recomendado que o faça 

```python
from sklearn import model_selection

from sklearn.model_selection import train_test_split

dataTrain, dataTest = train_test_split(dataFiltered, test_size=0.2)
# cria um modelo de validação cruzada
seed = 7
kfold = model_selection.KFold(n_splits=10, random_state=seed)
```

Cria os estimadores do nosso ensemble

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier

# separando os dados denovo
nColumns = dataTrain.shape[1] - 1
trainValues = dataTrain.values
trainFeatures = trainValues[:,0:nColumns-1]
trainTargets = trainValues[:,nColumns]
# cria os preditores
estimators = []
model1 = LogisticRegression()
estimators.append(('logistic', model1))
model2 = DecisionTreeClassifier()
estimators.append(('cart', model2))
model3 = SVC()
estimators.append(('svm', model3))
# cria o ensemble
ensemble = VotingClassifier(estimators)
ensemble.fit(trainFeatures, trainTargets)
results = model_selection.cross_val_score(ensemble, trainFeatures, trainTargets, cv=kfold)
print('Accuracy: {0:0.2f}'.format(results.mean()))
```

# Etapa 5: Realizar a classificação

```python
testValues = dataTest.values
testFeatures = testValues[:,0:nColumns-1]
testTargets = testValues[:,nColumns]

# realiza a predição sobre a base de teste
predictions = ensemble.predict(testFeatures)
```

# Etapa 6: Avaliar os resultados da predição

```python
# avalia o resultado 
from sklearn.metrics import accuracy_score
# realiza o calculo da acuracia
accuracy = accuracy_score(testTargets, predictions)
# exibe na tela o resultado final
print('Accuracy: {0:0.2f}'.format(accuracy))
```
