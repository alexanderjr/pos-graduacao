# Etapa 4: Criar e treinar o modelo de classificação

Esse é o momento de escolher qual o algoritmo você quer usar. Usar mais de um (ou um método ensemble) ajuda a ter a noção de qual a melhor estratégia para o problema.

Esse também é o momento de criar as suas bases de treinamento, validação e teste. Realizar a validação cruzada também ajuda bastante e é muito recomendado que o faça 

```python
from sklearn import model_selection

from sklearn.model_selection import train_test_split

dataTrain, dataTest = train_test_split(dataFiltered, test_size=0.2)
# cria um modelo de validação cruzada
seed = 7
kfold = model_selection.KFold(n_splits=10, random_state=seed)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
metricsValues = []
metricsLabels = []
temp = []
metrics = []
```

# Tentativa 1 - Utilizando Ensemble

Cria os estimadores do nosso ensemble

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import VotingClassifier
import time

#iniciando tempo
start = time.clock()

# separando os dados denovo
nColumns = dataTrain.shape[1] - 1
trainValues = dataTrain.values
trainFeatures = trainValues[:,0:nColumns-1]
trainTargets = trainValues[:,nColumns]
# cria os preditores
estimators = []
model1 = LogisticRegression()
estimators.append(('logistic', model1))

model2 = DecisionTreeClassifier()
estimators.append(('cart', model2))

model3 = SVC()
estimators.append(('svm', model3))

# model4 = MultinomialNB()
# estimators.append(('nb', model4))

# cria o ensemble
ensemble = VotingClassifier(estimators)
ensemble.fit(trainFeatures, trainTargets)
results = model_selection.cross_val_score(ensemble, trainFeatures, trainTargets, cv=kfold)
print('Acurácia do Treinamento: {0:0.2f}'.format(results.mean()))
end = time.clock()
trainning_time_ensemble = end-start
```

# Tentativa 1: Realizar a classificação

```python
testValues = dataTest.values
testFeatures = testValues[:,0:nColumns-1]
testTargets = testValues[:,nColumns]

#iniciando tempo
start = time.clock()
# realiza a predição sobre a base de teste
predictions = ensemble.predict(testFeatures)
end = time.clock()

metricsLabels.append('Ensemble')
temp = []

# metricas - Ensemble
temp.append(accuracy_score(testTargets, predictions))
temp.append(precision_score(testTargets, predictions))
temp.append(recall_score(testTargets, predictions))
temp.append(trainning_time_ensemble)
temp.append(end-start)
metrics.append(temp)
```
